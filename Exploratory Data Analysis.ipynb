```python
 corrMat = credit.corr()
print(corrMat)
```

```python
covMat = credit.cov()
print(covMat)
```

```python
#plt.hist(credit,bins=25)
#plt.show()
credit.hist(bins=25, color='steelblue', edgecolor='black', linewidth=1.0,xlabelsize=8, ylabelsize=8, grid=False)    
plt.tight_layout(rect=(0, 0, 2, 2))  
```

```python
f, ax = plt.subplots(figsize=(20, 10))
hm = sns.heatmap(round(corrMat,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Credit Attributes Correlation Heatmap', fontsize=14)
```

```python
# Pair-wise Scatter Plots
cols = ['PAY_0', 'PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']
pp = sns.pairplot(credit[cols], height=2, aspect=2,
                  plot_kws=dict(edgecolor="k", linewidth=1),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.5)
t = fig.suptitle('Credit Past Payment Attributes Pairwise Plots', fontsize=14)
```

```python
cols = ['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
pp = sns.pairplot(credit[cols], height=2, aspect=2,
                  plot_kws=dict(edgecolor="k", linewidth=1),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.4)
t = fig.suptitle('Credit Previous Payment Attributes Pairwise Plots', fontsize=14)
```

```python
cols = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']
pp = sns.pairplot(credit[cols], height=2, aspect=2,
                  plot_kws=dict(edgecolor="k", linewidth=1),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.4)
t = fig.suptitle('Credit Bill Amount Attributes Pairwise Plots', fontsize=14)
```
